```python
import nltk
import matplotlib
import matplotlib.pyplot as plt
from nltk.probability import FreqDist

with open("avatarSpeeches.txt", 'r', encoding='utf-8') as file:
    textRaw = file.read()
    tokens = textRaw.split()
    
    tokens = [w.lower() for w in tokens]

    textObj = nltk.Text(tokens)

    # Most common words:
    dist = FreqDist(tokens)
    print("10 most common words:", dist.most_common(10))

    # Length of text
    entire = len(textObj)
    print("Combined text length:", entire)

    # Words in similar contexts as avatar
    print("Words in similar contexts as 'avatar':")
    textObj.similar("avatar")

    # Plot
    words = ["aang", "sokka", "zuko", "katara", "toph"]
    nltk.draw.dispersion_plot(textObj, words)
```
