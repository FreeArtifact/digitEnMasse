3b. Prepare a code block that depends on running the previous code block and then reads from each directory of files and calculates what's needed to indicate **lexical diversity**.
    Running the cell that you prepare should reveal the lexical diversity of each set of files. *(3 pts)*

````python
def lexicalDiversity(token_list):
    return len(set(token_list)) / len(token_list)


# File number one: Mulholland
mul_script = open("text-file_1/mul-scrape.txt", "r").read()
tokens = mul_script.split()
#normalize
mul_tokens = [w.lower() for w in tokens]

diversity = lexicalDiversity(mul_tokens)
print("==Mulholland Script==")
print(f"Total tokens:      {len(mul_tokens)}")
print(f"Unique tokens:     {len(set(mul_tokens))}")
print(f"Lexical diversity: {diversity:.4f}\n")

#================================================
# File number two: Environmental
environ_text = open("text-file_2/og-memo-scrape.txt", "r").read()
tokens = environ_text.split()
#normalize
environ_tokens = [w.lower() for w in tokens]

diversity = lexicalDiversity(environ_tokens)
print("==Environmental Text==")
print(f"Total tokens:      {len(environ_tokens)}")
print(f"Unique tokens:     {len(set(environ_tokens))}")
print(f"Lexical diversity: {diversity:.4f}\n")
````
